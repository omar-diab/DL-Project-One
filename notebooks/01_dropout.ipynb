# 01 - Dropout

This notebook evaluates dropout on the same CIFAR-10 baseline setup.

## Goal
The goal is to test whether dropout reduces overfitting and improves validation/test performance under the same controlled setup.

## Fixed Conditions
- same fixed split
- same MLP architecture
- same optimizer
- same learning rate
- same epoch limit
- same seed

Only dropout is enabled in this notebook.

```python
# Import shared data loading, training, and plotting utilities
from src.data import get_dataloaders
from src.train_eval import run_experiment
from src.plots import plot_all_curves
```

```python
# Load the fixed train / validation / test split
train_loader, val_loader, test_loader = get_dataloaders()
```

```python
# Run the dropout experiment under the same controlled setup
dropout_result = run_experiment(
    train_loader=train_loader,
    val_loader=val_loader,
    test_loader=test_loader,
    experiment_name="Dropout",
    hidden_sizes=[1024, 512, 256],
    use_dropout=True,
    dropout_p=0.5,
    use_batchnorm=False,
    use_early_stopping=False,
    learning_rate=0.01,
    weight_decay=0.0,
    use_explicit_l2=False,
    max_epochs=50,
    seed=42
)
```

```python
# Plot loss, accuracy, and generalization gap curves
plot_all_curves(dropout_result, save_dir="results/figures", show=True)
```
