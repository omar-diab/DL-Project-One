```python
import os
import sys

repo_root = os.getcwd()

if repo_root not in sys.path:
    sys.path.append(repo_root)

src_path = os.path.join(repo_root, "src")
if src_path not in sys.path:
    sys.path.append(src_path)

print("Repo root:", repo_root)
print("src path:", src_path)
```

```python
from src.data import get_dataloaders
from src.train_eval import run_experiment
from src.plots import plot_all_curves, make_summary_table, save_summary_table

train_loader, val_loader, test_loader = get_dataloaders()

common_kwargs = {
    "train_loader": train_loader,
    "val_loader": val_loader,
    "test_loader": test_loader,
    "hidden_sizes": [1024, 512, 256],
    "learning_rate": 0.01,
    "max_epochs": 50,
    "seed": 42,
}

experiment_configs = [
    {
        "key": "baseline",
        "experiment_name": "Baseline",
        "use_dropout": False,
        "use_batchnorm": False,
        "use_early_stopping": False,
        "weight_decay": 0.0,
        "use_explicit_l2": False,
    },
    {
        "key": "dropout",
        "experiment_name": "Dropout",
        "use_dropout": True,
        "dropout_p": 0.5,
        "use_batchnorm": False,
        "use_early_stopping": False,
        "weight_decay": 0.0,
        "use_explicit_l2": False,
    },
    {
        "key": "early_stopping",
        "experiment_name": "EarlyStopping",
        "use_dropout": False,
        "use_batchnorm": False,
        "use_early_stopping": True,
        "patience": 5,
        "min_delta": 1e-4,
        "weight_decay": 0.0,
        "use_explicit_l2": False,
    },
    {
        "key": "batchnorm",
        "experiment_name": "BatchNorm",
        "use_dropout": False,
        "use_batchnorm": True,
        "use_early_stopping": False,
        "weight_decay": 0.0,
        "use_explicit_l2": False,
    },
    {
        "key": "dropout_batchnorm",
        "experiment_name": "Dropout + BatchNorm",
        "use_dropout": True,
        "dropout_p": 0.5,
        "use_batchnorm": True,
        "use_early_stopping": False,
        "weight_decay": 0.0,
        "use_explicit_l2": False,
    },
    {
        "key": "l2_dropout",
        "experiment_name": "L2 (MAP Gaussian Prior) + Dropout",
        "use_dropout": True,
        "dropout_p": 0.5,
        "use_batchnorm": False,
        "use_early_stopping": False,
        "weight_decay": 1e-4,
        "use_explicit_l2": True,
    },
]

results = {}

for cfg in experiment_configs:
    cfg = cfg.copy()
    key = cfg.pop("key")
    print(f"\n{'='*80}")
    print(f"Running: {cfg['experiment_name']}")
    print(f"{'='*80}")

    result = run_experiment(**common_kwargs, **cfg)
    results[key] = result
    plot_all_curves(result, save_dir="results/figures", show=True)

summary_df = make_summary_table(results)
save_summary_table(summary_df, "results/tables/experiment_summary.csv")

display(summary_df)
```
