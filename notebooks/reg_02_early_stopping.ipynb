# 02 - Early Stopping

This notebook evaluates early stopping on the same CIFAR-10 baseline setup.

## Goal
The goal is to stop training when validation loss no longer improves, and to observe whether this reduces overfitting and improves generalization.

## Fixed Conditions
- same fixed split
- same MLP architecture
- same optimizer
- same learning rate
- same epoch limit
- same seed

Only early stopping is enabled in this notebook.

```python
# Import shared data loading, training, and plotting utilities
from src.data import get_dataloaders
from src.train_eval import run_experiment
from src.plots import plot_all_curves
```

```python
# Load the fixed train / validation / test split
train_loader, val_loader, test_loader = get_dataloaders()
```

```python
# Run the early stopping experiment (stops training based on validation loss)
early_result = run_experiment(
    train_loader=train_loader,
    val_loader=val_loader,
    test_loader=test_loader,
    experiment_name="EarlyStopping",
    hidden_sizes=[1024, 512, 256],
    use_dropout=False,
    use_batchnorm=False,
    use_early_stopping=True,
    patience=5,
    min_delta=1e-4,
    learning_rate=0.01,
    weight_decay=0.0,
    use_explicit_l2=False,
    max_epochs=50,
    seed=42
)
```

```python
# Plot loss, accuracy, and generalization gap curves
plot_all_curves(early_result, save_dir="results/figures", show=True)
```

```python
# Print best validation-based stopping point
print("Best epoch:", early_result.get("best_epoch"))
print("Best val loss:", early_result.get("best_val_loss"))
```
