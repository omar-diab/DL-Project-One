# 04 - Dropout + BatchNorm

This notebook evaluates the combination of dropout and batch normalization on the same CIFAR-10 baseline setup.

## Goal
The goal is to see whether combining dropout and batch normalization gives a stronger regularization effect than using them separately.

## Fixed Conditions
- same fixed split
- same MLP architecture
- same optimizer
- same learning rate
- same epoch limit
- same seed

Only dropout and batch normalization are enabled in this notebook.

```python
# Import shared data loading, training, and plotting utilities
from src.data import get_dataloaders
from src.train_eval import run_experiment
from src.plots import plot_all_curves
```

```python
# Load the fixed train / validation / test split
train_loader, val_loader, test_loader = get_dataloaders()
```

```python
# Run the combined dropout + batch normalization experiment
dropout_bn_result = run_experiment(
    train_loader=train_loader,
    val_loader=val_loader,
    test_loader=test_loader,
    experiment_name="Dropout + BatchNorm",
    hidden_sizes=[1024, 512, 256],
    use_dropout=True,
    dropout_p=0.5,
    use_batchnorm=True,
    use_early_stopping=False,
    learning_rate=0.01,
    weight_decay=0.0,
    use_explicit_l2=False,
    max_epochs=50,
    seed=42
)
```

```python
# Plot loss, accuracy, and generalization gap curves
plot_all_curves(dropout_bn_result, save_dir="results/figures", show=True)
```
