# 05 - L2 (MAP Gaussian Prior) + Dropout

This notebook evaluates dropout together with L2 regularization, interpreted as MAP estimation under a Gaussian prior.

## Goal
The goal is to explicitly show the MAP interpretation of L2 regularization in practice.

## Theory
We use the following objective:

$$
\mathcal{L}_{MAP} = \mathcal{L}_{CE} + \lambda \sum_i w_i^2
$$

This corresponds to maximum a posteriori (MAP) estimation under a zero-mean isotropic Gaussian prior on the weights:

$$
p(\mathbf{w}) = \mathcal{N}(0, \sigma^2 I)
$$

Here:
- `CrossEntropyLoss` corresponds to the negative log-likelihood
- the explicit L2 term corresponds to the negative log-prior
